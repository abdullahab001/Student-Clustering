# Manual Clustering on the basis of learner’s company, job position and years of experience
* Getting the 5 point summary of CTC (mean, median, max, min, count etc) on the basis of Company, Job Position, Years of Experience
* Merging the same with original dataset carefully and creating some flags showing learners with CTC greater than the Average of their Company’s department having same Years of Experience - Call that flag designation with values [1,2,3]
* Doing above analysis at Company & Job Position level. Name that flag Class with values [1,2,3]
* Repeating the same analysis at the Company level. Name that flag Tier with values [1,2,3]


```python
ctc_25 = df_temp.groupby(['company_hash', 'job_position', 'years_of_exp'])['ctc'].quantile(0.25).reset_index()
ctc_50 = df_temp.groupby(['company_hash', 'job_position', 'years_of_exp'])['ctc'].quantile(0.50).reset_index()
ctc_75 = df_temp.groupby(['company_hash', 'job_position', 'years_of_exp'])['ctc'].quantile(0.75).reset_index()
ctc_count = df_temp.groupby(['company_hash', 'job_position', 'years_of_exp'])['ctc'].count().reset_index()

ctc_df = ctc_25.merge(ctc_50, on=['company_hash', 'job_position', 'years_of_exp'], suffixes=('_25', '_50'))
ctc_df = ctc_df.merge(ctc_75, on=['company_hash', 'job_position', 'years_of_exp'])
ctc_df = ctc_df.merge(ctc_count, on=['company_hash', 'job_position', 'years_of_exp'], suffixes=('_75', '_count'))

# Correcting the column name suffix for count
ctc_df.rename(columns={'ctc': 'ctc_count'}, inplace=True)

ctc_df.shape
```

# Data processing for Unsupervised clustering - Label encoding/ One- hot encoding, Standardization of data

```python
scaler
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_features = ['designation', 'class', 'tier']

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(scaler[categorical_features])
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out(categorical_features))
df_encoded = pd.concat([scaler.drop(columns=categorical_features), encoded_df], axis=1)
#df_encoded
```

```python
import pandas as pd

def frequency_encoding(df, columns):
    for col in columns:
        freq_encoding = df[col].value_counts(normalize=True)
        df[col + '_freq_enc'] = df[col].map(freq_encoding)
    return df

# Example usage
# Assuming df is your DataFrame and cat_columns is a list of categorical columns
cat_columns = ['company_hash', 'job_position']
frq_encoded = frequency_encoding(df_encoded, cat_columns)
frq_encoded
```

```python
frq_encoded[frq_encoded['company_hash_freq_enc'].round(6) == 0.727181]
```

```python
X = frq_encoded[['designation_1',
       'designation_2', 'designation_3', 'class_1', 'class_2', 'class_3',
       'tier_1', 'tier_2', 'tier_3', 'ctc', 'years_of_exp',
       'company_hash_freq_enc', 'job_position_freq_enc']].copy()
X.rename(columns={'company_hash_freq_enc': 'company_hash','job_position_freq_enc':'job_position'}, inplace=True)
X = X.drop_duplicates()
X = X[X['years_of_exp'] != 124]
```

Average age of humans is 70+ but still I left the records with experience as 54 but 124 years of experince won't make any sense.

```python
from sklearn.preprocessing import StandardScaler
X = X.reset_index(drop=True)
scale = StandardScaler()
num_columns = ['ctc','years_of_exp']
scaled_ftrs = scale.fit_transform(X[num_columns])
scaled_df = pd.DataFrame(scaled_ftrs, columns=num_columns)
data = pd.concat([X, scaled_df], axis=1)
data
```

# Unsupervised Learning - Clustering
* Checking clustering tendency
* Elbow method
* K-means clustering
* Hierarchical clustering (you can do this on a sample of the dataset if your process is taking time)


```python
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
distortions = []
max_clusters = 10
for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=0)
    kmeans.fit(data)
    distortions.append(kmeans.inertia_)
```

```python
plt.plot(range(1, max_clusters + 1), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=0)

# Fit KMeans to your data
kmeans.fit(data.values)

# Get cluster labels and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters (optional)
plt.scatter(data.values[:, 0], data.values[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=100, color='red')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
import numpy as np

def silhouette_score_kmeans(X, k):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    return silhouette_score(X, kmeans.labels_)

for k in range(2, 6):
    print(f"Silhouette Score for {k} clusters:", silhouette_score_kmeans(data, k))
```
